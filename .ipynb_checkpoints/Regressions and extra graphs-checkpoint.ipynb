## Logistic regression to see the crimes of elderly in prison and how it correlates to release type (Either Conditional or Unconditional)

```python
import numpy as np
from patsy import dmatrices
import statsmodels.api as sm
import sklearn.metrics as metrics
import sklearn.model_selection as cross_validation
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Organize data so we have only what we want (Released as elderly, in the year 2013)
dfcopy = df3.copy()
dfcopy = dfcopy.query('AGERLSE == "55+ years"')
dfcopy = dfcopy.loc[dfcopy['RELYR'] == 2013]
print("size of DF: ", len(dfcopy))


dfcopy.head()
```

```python
conditionalReleaseList = []
indexToDrop = []
indexCount = -1
#Iterate through our DF and construct a list called "conditionalReleaseList", which is either 1 or 0
for index, row in dfcopy.iterrows():
    indexCount = indexCount + 1
    if row["RELTYPE"] == "Conditional release":
        conditionalReleaseList.append(1.0)
    elif row["RELTYPE"] == "Unconditional release":
        conditionalReleaseList.append(0.0)
    else:
        indexToDrop.append(indexCount)


#Drop the indices we don't want to consider (missing values and deaths)
dfcopy = dfcopy.drop(dfcopy.index[indexToDrop])


print("Size Check")
print("Size of DF: ", len(dfcopy))
print("Size of conditionalReleaseList: ", len(conditionalReleaseList) )
dfcopy["CONDITIONAL_RELEASE"] = conditionalReleaseList
        
dfcopy.head()
```

```python
#In our analysis, we only want to consider parts of the DF that would be relevant to a prisoners release.
#We wil drop information we do not want to consider (irrelevant information AND information that is the same for everyone)

regressionSet = dfcopy[["EDUCATION", "OFFDETAIL", "RACE", "CONDITIONAL_RELEASE",]]

print("before NA drop: ", len(regressionSet))
regressionSet = regressionSet.dropna()
print("after NA drop: ", len(regressionSet))

regressionSet.tail()
```

```python
#Create indicies to randomly decide which bits of data we will seperate from the regression
import random
numberkeep = .8 * len(regressionSet)
numberkeep = int(numberkeep)

randomindices = []
count = 0
print(numberkeep)
while count < numberkeep:
    randNumber = random.randint(0,(len(regressionSet) - 1))
    if randNumber in randomindices:
        continue
    else:
        randomindices.append(randNumber)
        count = count + 1
        
        
        
#Checking to make sure that all the elements are unqiue
def checkUnique(alist):
    if len(set(alist)) == len(alist):
        print ("Unique list!")
    else:
        print(set(alist))
        print(len(set(alist)))
        print("Not Unique!")
        
print(randomindices)
print(len(randomindices))
checkUnique(randomindices)
```

```python
# Randomly seperate 20% of the data so we can check our regression against it.
print("# of rows of data: ", len(regressionSet))
print()
trainset = regressionSet.iloc[randomindices]
testset = regressionSet.drop(regressionSet.index[randomindices])
print("# rows to drop:", len(randomindices))
print("# of rows of data after drop left in testset: ", len(testset))
```

```python
#Prepare X and y test for checking the regression

Xtest = pd.get_dummies(testset)

Xtest = Xtest.drop(['EDUCATION_Ungraded/unknown', 'OFFDETAIL_Missing', 'RACE_Missing' ], axis = 1)

#EDUCATION_<HS diploma/GED has issues showing in regression because of name.

Xtest.rename(columns = {'EDUCATION_<HS diploma/GED':'EDUCATION_Under HS diploma/GED'}, inplace = True)

ytest = Xtest["CONDITIONAL_RELEASE"]

Xtest = Xtest.drop(['CONDITIONAL_RELEASE'], axis=1)

intercept = [1.0] * len(ytest)

Xtest["Intercept"] = intercept

Xtest.head()
```

```python

# Code to evaluate regression borrowed from Professor Smith's lecture
def evaluate(y_train, X_train, y_test, X_test, threshold):

    # learn model on training data
    logit = sm.Logit(y_train, X_train)
    result = logit.fit(maxiter=20000, missing="drop", method='nm')
    
    # make probability predictions on test data
    y_pred = result.predict(X_test)
    
    # threshold probabilities to create classifications
    y_pred = y_pred > threshold
    
    # report metrics
    precision = metrics.precision_score(y_test, y_pred)
    recall = metrics.recall_score(y_test, y_pred)
    return precision, recall

precision, recall = evaluate(ytrain, Xtrain, ytest, Xtest, 0.5)

print('\nPrecision: {:0.3f}, Recall: {:0.3f}'.format(precision,recall))
```

```python
#ROC, taken from Smith's lecture

def evaluate_ROC(y_train, X_train, y_test, X_test):

    # learn model on training data
    logit = sm.Logit(y_train, X_train)
    result = logit.fit(maxiter=20000, missing="drop", method='nm')
    
    # make probability predictions on test data
    y_pred = result.predict(X_test)
    
    # build ROC 
    fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
    roc_auc = metrics.auc(fpr, tpr)

    return fpr, tpr, thresholds, roc_auc

fpr, tpr, thresholds, roc_auc = evaluate_ROC(ytrain, Xtrain, ytest, Xtest)
```

```python
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
```
